# 小白的 Github 入门手册 —— 社交平台？

>  市面上的笔记软件，社交平台非常多了，但是用着总觉得哪里不够顺手，直到我这个编程小白发现了Github。起因是年纪大了之后，发现生活中有很多方面需要记录，比如说学习，工作，炒股票还有一些兴趣爱好等等方面，一开始的时候我是选择用word进行记录，后来发现word不好用，有很多格式不支持，而且每次打开文件查找材料非常麻烦，后来发现了一款不错的笔记软件Notion，确实是可以系统化有逻辑的记录生活了，但是Notion不方便的地方在于不容易与其它人互动，而且过于依赖软件本身，万一哪天这个软件崩了，所有材料都找不到，想想都可怕。

## 为什么使用Github

方便系统性地记录学习和生活，方便归纳整理各类素材，而且可以跟他人进行互动，随着编程的普及说不定哪天就出圈火了。


## 注册

网址https://github.com，登陆上之后，点击右上角的图标，填写邮箱还有密码就可以注册了。很多人可能会卡在这一步，然后就对这篇文章失去了兴趣，b站上有很多可以快速打开Github的方法，感兴趣的小伙伴可以去搜一搜。

## Github的一些概念  

登陆后，点击右上角图标下的your profile，就可以进入主页。进入主页后，你会看到下面这张图。


左边这块是你的简介，包括你的名字，个人签名，定位等等，点击edit profile就可以编辑自己的个人信息。

上方有几个选项，overview 概览，在这里你可以看到自己的主页，主页上记录了你的操作，比如你在哪天有更新，你最新的仓库，以及你的操作等等。

Repositories 仓库，这里就是你创作的地方，你可以根据需要建立不同主题的仓库，比如美妆主题仓库，日常生活记录，等等方面。在仓库里面你可以建立很多的文件，大部分文件类型都可以接受，像我用的比较多的是jpynb，主要用来记录文字，图片，影像等等，不过大部分人还是保存代码的居多。除了记录之外，其它人可以在这个仓库里面的pull request 和 discussion 跟你进行互动。如果你不想让别人看到这个仓库，在建立的时候，你可以选择private，这样就你一个人可以看到了。

project 项目，这个一般是用来编程的，后续再更新。

stars 收藏，这个地方是用来收藏你看到的比较好的仓库。

## 如何在Github上创建文稿  

在我们建立了github的仓库之后，紧接着我们就会在仓库中上传文件。我上传的一般是jpynb，小白们可以跟我一样先从jpynb开始用起，这个东西其实跟word是比较相似的，但是它不仅可以编写文稿，而且可以编写代码，运行代码，做各种数据图等等。

### jpynb 文件的创建方法
>* 首先你需要先下载一个编译器，[Visual Studio Code](https://code.visualstudio.com/docs)，点击右上角的download
>* 然后下载我这篇文章作为模版，用Visual Studio Code 打开之后，你就可以编辑自己的内容了。直接编辑别人的文件，是最快上手的方法。
