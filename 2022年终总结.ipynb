# 2022年终总结

2022年对于我来说是无比重要的一年，这一年我想明白了很多的事情，以至于我从未觉得过去27年漫长时光里有哪一年像2022年一样重要。

2022年大部分时间都在封控中度过，也因此多了很多思考的时间，可以回顾自己过往的生活，并想明白人生中很多的事情。譬如

> * 我这辈子应该做什么
> * 时间的重要性，如何管理自己，对重要的事情投入时间
> * 成为别人需要的人，而不是一味的索取
> * 真实和幻想的区别
> * 生产的重要性
> * 积极与他人建立联系
> * 克服焦虑
> * 人生很长，每个人都有机会
> * 希望的重要性
> * 永远保持好奇心
> * 为自己打工而不是为别人打工
> * 不要生气
> * 系统化的重要性
> * ....

这一年里我想明白了很多方面的事情，看似是有很多方面，实际上想明白了最根本的问题大部分问题也就想明白了


## 时间就是一切

我从未像今天这样，发现时间的重要性。我们来到这个世界上，唯一拥有的只有时间，对于一般人来说，父母所能给予的部分对于人生的影响可以说是正负相抵吧。非常有钱就会限制你的发展，也就是所谓的资源诅咒，一般有钱的能给你的帮助也不多。

本质上来说**你拥有的只有时间**，不管是感情还是事业成就，又或者是你的快乐，实际上都是用你的时间换取的。不管你通过时间换取金钱再换取其他的，还是你直接用时间获得快乐，都是一样的。

正因为时间太过于宝贵，所以应该尽可能集中在某一块主要的事情上，才有可能做出一些事情来。

时间应该分块投放，比如说每周可以在固定时间跟父母联系，或者跟朋友联系，或者是固定的时间看书。如果注意力过于分散，大部分事情都做不成。

## 成为别人需要的人，而不是一味的索取

在过去很长一段时间里，我都想过一件事情，就是一劳永逸。这也许是原生家庭的影响。也许这个想法很多人都想过，上了好的大学未来就一定有希望，有一份工作，然后就等着别人给你发钱就结束了，或者找到对象，然后就可以开始摆烂了。事实上，这只不过是起点而已。在未来漫长的时光里，不停的成为对别人有用的人，从而换取事业，金钱还有感情。

成为对别人有用的人是一件很幸福的事情，一味的索取比如要求涨薪这些，其实并没有什么意义，如果你实在优秀而市场又是有效的，那你几乎不可能会吃亏。

## 真实与幻想的区别，对于爱情的了解

这是感情观上的感悟，在过去很长的一段时间里，我的感情观都很模糊的，全靠直觉。后来读了一些书，反思了一下。

发现自己过去大部分的感情都来自于想象，两个人之间也许并没有很深层次的了解。我们满足彼此的需求只是在满足彼此的想象，而想象并不会激励你的生活。也不会激励对方的生活。

在经历了一些事情，读了一些书之后，我对于爱的认识发生了很大的改变，在我心中真正的爱，是希望对方更好，而不是占有，如果在当下双方满足了彼此的需求走到了一起，那很好，但随着时间的推移，两个人如果已经无法走下去了，那应该祝福并尊重对方，不要强行转移他人的意志。

我们能做的只是当我们跟别人之间产生了爱情的连接时，如何经营这段爱情，并使之激励你的生活，是更重要的事情。

## 希望是很重要的东西

我想这是我这一年领悟到最重要的事情，现在这个年代，抱有希望可能是被大部分人认为很傻的事情。

我们寻找对象基于对方当下的条件，忽视对方身上所有的可能性，我们找工作更是如此，社会越来越注重当下的现实，并预设当下的现状将持续下去，我们用所谓的阶级固化以及日韩的现状来解释这一切，但我们忘记了变化是世上唯一不变的事。哪怕未来的现实确如你所想的悲观，不抱有希望的活着也是更次的选择。
